```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.manifold import SpectralEmbedding
import matplotlib
from mpl_toolkits.mplot3d import Axes3D

#### The clustering Algorithms ####
from cluster import AffinityProp_BestSil
from cluster import AgglomerativeClustering_BestSil
from cluster import DBSCANClustering_BestSil
from cluster import MiniBatch_BestSil
from cluster import GMM_BestSil
from cluster import AffinityProp_BestAri
from cluster import AgglomerativeClustering_BestAri
from cluster import DBSCANClustering_BestAri
from cluster import MiniBatch_BestAri
from cluster import GMM_BestAri
```

#### Load the Iris Data set as sample data.
The dataset is located here: https://archive.ics.uci.edu/ml/datasets/iris
The data (matrix of size n x d or 150 x 4) is output to a text file - outTestMat.txt
The file is to be loaded into Urerf and run to produce a similarity matrix (150x150)

```python
iris = datasets.load_iris()
numFeatures = 4
X = iris.data[:, :numFeatures]
mat = []
for i in range(0, len(X)):
    mat.append(X[i])
mat = np.matrix(mat, dtype = float)

# #############################################################################
# Print to file to run in Urerf

with open('../output/outTestMat.txt','wb') as f:
    for line in mat:
        np.savetxt(f, line, fmt='%.2f')           
print("Feature matrix of shape", mat.shape, "outputted. Run Urerf in R now.")
```

### Note the default parameters:
        - num trees = 500
        - k = 10
        - depth = 4
        - dim = 2

### Plotting functions

```python
###### Plotting ######
def graph2DLabels(mat, labels):
    fig = plt.figure(3, figsize=(10, 10))
    ax = plt.Axes(fig, rect=[0, 0, .5, .5])
    z = list(labels)
    N = len(mat[:,1])
    x = [mat[:,0]]
    y = [mat[:,1]]
    cmap = matplotlib.cm.get_cmap('plasma')
    normalize = matplotlib.colors.Normalize(vmin=min(z), vmax=max(z))
    colors = [cmap(normalize(value)) for value in z]
    plt.scatter(x, y, color = colors)
    plt.title("Iris Dataset")
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    
def graph3DLabels(mat, labels):
    fig = plt.figure(4, figsize=(10, 10))
    ax = Axes3D(fig, rect=[0, 0, .5, .5], elev=40, azim=150)
    ax.scatter(mat[:, 0], mat[:, 1], mat[:, 2], c=labels, cmap = "plasma")
    ax.set_title("Iris Dataset")
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')
    
def plotMultipleDims(dataMatrix, colors, dim, trial_num):
    for i in range(trial_num * len(dim)):
        mat_ = np.matrix(np.vstack(dataMatrix[i][1]))
        dimension = dataMatrix[i][0]
        if i%trial_num == 0:
            plt.scatter(mat_[:,0].tolist(), mat_[:,1].tolist(), c = colors[dim.index(dimension)], label=dimension)
        else:
            plt.scatter(mat_[:,0].tolist(), mat_[:,1].tolist(), c = colors[dim.index(dimension)])
def plotReg(dataMatrices, labels, ylab, xlab, colors):
    for i in range(len(dataMatrices)):
        dat = dataMatrices[i]
        plt.scatter(dat[:,0].tolist(), dat[:,1].tolist(), c = colors[i], label=labels[i])
    plt.ylabel(ylab)
    plt.xlabel(xlab)
```

### Plot the iris data using the first two dimension and first 3 dimensions.

```python
plt.figure(1)
graph2DLabels(mat, iris.target)
plt.figure(2)
graph3DLabels(mat, iris.target)
```

### This code loads the similarity matrix produced from Urerf, embeds it into a set number of dimensions, and clusters it using GMM and Mini Batch K Means
### It also looks how how varying the dimensionality affects the number of clusters.

```python
# Set parameters for the test #
trial_num = 50
t = float(trial_num)
n = 2

# Load the similarity matrix #
fname = '../input/inTestMat.txt'
matrix = open(fname, 'r').read()
matrix = [item.split() for item in matrix.split('\n')[:-1]]
inputMat = np.matrix(matrix, dtype = float)
embedding = SpectralEmbedding(n_components=n, affinity = 'precomputed')
inputMat_transformed = embedding.fit_transform(inputMat)

# Lists to hold all of the trials #
gmm_data_sil = []
mini_data_sil = []
gmm_data_ari = []
mini_data_ari = []

# Call clustering functions which vary the number of clusters from 1 to 20 #
for i in range(trial_num):
    bestClustGMM, maxSilGMM, trials_GMM = GMM_BestSil(inputMat_transformed, clustInit = 1, clustInc = 1, clustMax = 20)
    bestClustMini, maxSilMini, trials_Mini = MiniBatch_BestSil(inputMat_transformed, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
    gmm_data_sil.append(trials_GMM)
    mini_data_sil.append(trials_Mini)
    bestClustGMM, maxSilGMM, trials_GMM = GMM_BestAri(inputMat_transformed, iris.target, clustInit = 1, clustInc = 1, clustMax = 20)
    bestClustMini, maxSilMini, trials_Mini = MiniBatch_BestAri(inputMat_transformed, iris.target, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
    gmm_data_ari.append(trials_GMM)
    mini_data_ari.append(trials_Mini)

# Varying the embedding dimension #
trial_num = 10
dim =[2, 3, 4, 5, 10, 15]

# Lists to hold all of the trials #
gmm_data_sil_dim = []
mini_data_sil_dim = []
gmm_data_ari_dim = []
mini_data_ari_dim = []

for d in dim:
    # Call clustering functions which vary the number of clusters from 1 to 20 #
    for i in range(trial_num):
        embedding = SpectralEmbedding(n_components=d, affinity = 'precomputed')
        inputMat_transformed = embedding.fit_transform(inputMat)
        bestClustGMM, maxSilGMM, trials_GMM = GMM_BestSil(inputMat_transformed, clustInit = 1, clustInc = 1, clustMax = 20)
        bestClustMini, maxSilMini, trials_Mini = MiniBatch_BestSil(inputMat_transformed, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
        gmm_data_sil_dim.append([d, trials_GMM])
        mini_data_sil_dim.append([d, trials_Mini])
        bestClustGMM, maxSilGMM, trials_GMM = GMM_BestAri(inputMat_transformed, iris.target, clustInit = 1, clustInc = 1, clustMax = 20)
        bestClustMini, maxSilMini, trials_Mini = MiniBatch_BestAri(inputMat_transformed, iris.target, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
        gmm_data_ari_dim.append([d, trials_GMM])
        mini_data_ari_dim.append([d, trials_Mini])
```

### Plot all results

```python
mini_data_ari = np.matrix(np.vstack(mini_data_ari))
gmm_data_ari = np.matrix(np.vstack(gmm_data_ari))
mini_data_sil = np.matrix(np.vstack(mini_data_sil))
gmm_data_sil = np.matrix(np.vstack(gmm_data_sil))

plt.figure(figsize=(25, 20))
plt.subplot(3,2,1)
plotReg([mini_data_ari, gmm_data_ari], ["Mini Batch K Means",'Gaussian Mixture'],'Adjusted Rand Index','Number of Clusters', ['y','b'])
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(3,2,2)
plotReg([mini_data_sil, gmm_data_sil], ["Mini Batch K Means",'Gaussian Mixture'],'Silhouette Coefficient','Number of Clusters', ['y','b'])
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(loc=9, bbox_to_anchor=(1.2, 0.6))

# Get a color map for the differing dimensions #
cmap = matplotlib.cm.get_cmap('plasma')
normalize = matplotlib.colors.Normalize(vmin=min(dim[0:3]), vmax=max(dim[0:3]))
colors = [cmap(normalize(value)) for value in dim[0:3]]
cmap = matplotlib.cm.get_cmap('rainbow')
normalize = matplotlib.colors.Normalize(vmin=min(dim[3:6]), vmax=max(dim[3:6]))
colors += [cmap(normalize(value)) for value in dim[3:6]]

plt.subplot(3,2,3)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Mini Batch K-Means")
plotMultipleDims(mini_data_ari_dim, colors, dim, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(3,2,4)
plt.ylabel('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.title("Mini Batch K-Means")
plotMultipleDims(mini_data_sil_dim, colors, dim, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "dimensions", loc=9, bbox_to_anchor=(1.2, 0.6))


plt.subplot(3,2,5)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Gaussian Mixture")
plotMultipleDims(gmm_data_ari_dim, colors, dim, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(3,2,6)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Gaussian Mixture")
plotMultipleDims(gmm_data_sil_dim, colors, dim, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "dimensions", loc=9, bbox_to_anchor=(1.2, 0.6))

```

### This code loads the similarity matrix produced from Urerf and clusters it using affinity propagation. It also translates the matrix into a distance matrix and clusters with DBSCAN and agglomerative clustering.

```python
# Set parameters for the test #
trial_num = 15
t = float(trial_num)
n = 2

# Load the similarity matrix #
fname = '../input/inTestMat.txt'
matrix = open(fname, 'r').read()
matrix = [item.split() for item in matrix.split('\n')[:-1]]
inputMat = np.matrix(matrix, dtype = float)
inputMat_transformed = 1 - (inputMat/inputMat.max())

# Lists to hold all of the trials #
af_data_sil = []
ag_data_sil = []
db_data_sil = []
af_data_ari = []
ag_data_ari = []
db_data_ari = []

# Call clustering functions which vary the number of clusters from 1 to 20 or the preference and epsilon parameters #
# which vary from their initial values below to the maximum element in each matrix #
for i in range(trial_num):
    maxPre, maxSilAf, trials_Af = AffinityProp_BestSil(inputMat, damp = .9, preferenceInit = 0.0, preferenceInc = .05)
    bestClustAg, maxSilAg, trials_Ag = AgglomerativeClustering_BestSil(inputMat_transformed, clustInit = 2, clustInc = 1, clustMax = 20)
    maxEps, maxSilDb, trials_Db = DBSCANClustering_BestSil(inputMat_transformed, epsInit = 0, epsInc = 0.05)
    af_data_sil.append(trials_Af)
    ag_data_sil.append(trials_Ag)
    db_data_sil.append(trials_Db)
    
    maxPre, maxAriAf, trials_Af = AffinityProp_BestAri(inputMat, iris.target, damp = .9, preferenceInit = 0.0, preferenceInc = .05)
    bestClustAg, maxAriAg, trials_Ag = AgglomerativeClustering_BestAri(inputMat_transformed, iris.target, clustInit = 2, clustInc = 1, clustMax = 20)
    maxEps, maxAriDb, trials_Db = DBSCANClustering_BestAri(inputMat_transformed, iris.target, epsInit = 0, epsInc = 0.05)
    af_data_ari.append(trials_Af)
    ag_data_ari.append(trials_Ag)
    db_data_ari.append(trials_Db)
    
```

### Plot all results

```python
af_data_ari = np.matrix(np.vstack(af_data_ari))
ag_data_ari = np.matrix(np.vstack(ag_data_ari))
db_data_ari = np.matrix(np.vstack(db_data_ari))
af_data_sil = np.matrix(np.vstack(af_data_sil))
ag_data_sil = np.matrix(np.vstack(ag_data_sil))
db_data_sil = np.matrix(np.vstack(db_data_sil))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plotReg([af_data_ari, db_data_ari, ag_data_ari], ["Affinity Propagation",'DBSCAN', 'Agglomerative'],'Adjusted Rand Index','Number of Clusters', ['y','b', 'g'])
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(1,2,2)
plotReg([af_data_sil, db_data_sil, ag_data_sil], ["Affinity Propagation",'DBSCAN', 'Agglomerative'],'Silhouette Coefficient','Number of Clusters', ['y','b','g'])
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(loc=9, bbox_to_anchor=(1.2, 0.6))
```

### The next test demonstrats the effect of varying the k parameter in Urerf on each algorithm that has achieved clusters thus far. 
Dbscan is not included because it fails to get over 1 cluster. 

```python
files = ["inTestMat.txt","inTestMatA.txt","inTestMatB.txt","inTestMatC.txt","inTestMatD.txt","inTestMatE.txt","inTestMatF.txt"]

# Varying the k dimension #
trial_num = 10
kVals =[10, 12, 15, 30, 50, 100, 120]

# Lists to hold all of the trials #
gmm_data_sil_k = []
mini_data_sil_k = []
gmm_data_ari_k = []
mini_data_ari_k = []
af_data_sil_k = []
ag_data_sil_k = []
af_data_ari_k = []
ag_data_ari_k = []
    

for i in range(len(files)):
    # Load the similarity matrix #
    fpath = '../input/' + files[i]
    matrix = open(fpath, 'r').read()
    matrix = [item.split() for item in matrix.split('\n')[:-1]]
    inputMat = np.matrix(matrix, dtype = float)
    inputMat_transformed = 1 - (inputMat/inputMat.max())
    embedding = SpectralEmbedding(n_components=2, affinity = 'precomputed')
    inputMat_embedded = embedding.fit_transform(inputMat)
    
    k = kVals[i]
    # Call clustering functions which vary the number of clusters from 1 to 20 #
    for i in range(trial_num):
        
        
        bestClustGMM, maxSilGMM, trials_GMM = GMM_BestSil(inputMat_embedded, clustInit = 1, clustInc = 1, clustMax = 20)
        bestClustMini, maxSilMini, trials_Mini = MiniBatch_BestSil(inputMat_embedded, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
        gmm_data_sil_k.append([k, trials_GMM])
        mini_data_sil_k.append([k, trials_Mini])
        
        bestClustGMM, maxAriGMM, trials_GMM = GMM_BestAri(inputMat_embedded, iris.target, clustInit = 1, clustInc = 1, clustMax = 20)
        bestClustMini, maxAriMini, trials_Mini = MiniBatch_BestAri(inputMat_embedded, iris.target, clustInit = 1, clustInc = 1, batchSize = 75, clustMax = 20)
        gmm_data_ari_k.append([k, trials_GMM])
        mini_data_ari_k.append([k, trials_Mini])
        
        maxPre, maxSilAf, trials_Af = AffinityProp_BestSil(inputMat, damp = .9, preferenceInit = 0.0, preferenceInc = .05)
        bestClustAg, maxSilAg, trials_Ag = AgglomerativeClustering_BestSil(inputMat_transformed, clustInit = 2, clustInc = 1, clustMax = 20)
        af_data_sil_k.append([k,trials_Af])
        ag_data_sil_k.append([k,trials_Ag])

        maxPre, maxAriAf, trials_Af = AffinityProp_BestAri(inputMat, iris.target, damp = .9, preferenceInit = 0.0, preferenceInc = .05)
        bestClustAg, maxAriAg, trials_Ag = AgglomerativeClustering_BestAri(inputMat_transformed, iris.target, clustInit = 2, clustInc = 1, clustMax = 20)
        af_data_ari_k.append([k,trials_Af])
        ag_data_ari_k.append([k,trials_Ag])
```

### Plot the results

```python
# Get a color map for the differing dimensions #
cmap = matplotlib.cm.get_cmap('summer')
normalize = matplotlib.colors.Normalize(vmin=min(kVals[0:3]), vmax=max(kVals[0:3]))
colors = [cmap(normalize(value)) for value in kVals[0:3]]
cmap = matplotlib.cm.get_cmap('rainbow')
normalize = matplotlib.colors.Normalize(vmin=min(kVals[3:7]), vmax=max(kVals[3:7]))
colors += [cmap(normalize(value)) for value in kVals[3:7]]
plt.figure(figsize=(25, 30))


plt.subplot(5,2,1)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Mini Batch K-Means")
plotMultipleDims(mini_data_ari_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(5,2,2)
plt.ylabel('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.title("Mini Batch K-Means")
plotMultipleDims(mini_data_sil_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "k-value", loc=9, bbox_to_anchor=(1.2, 0.6))


plt.subplot(5,2,3)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Gaussian Mixture")
plotMultipleDims(gmm_data_ari_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(5,2,4)
plt.ylabel('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.title("Gaussian Mixture")
plotMultipleDims(gmm_data_sil_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "k-value", loc=9, bbox_to_anchor=(1.2, 0.6))

plt.subplot(5,2,5)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Affinity Propagation")
plotMultipleDims(af_data_ari_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(5,2,6)
plt.ylabel('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.title("Affinity Propagation")
plotMultipleDims(af_data_sil_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "k-value", loc=9, bbox_to_anchor=(1.2, 0.6))

plt.subplot(5,2,7)
plt.ylabel('Adjusted Rand Index')
plt.xlabel('Number of Clusters')
plt.title("Agglomerative")
plotMultipleDims(ag_data_ari_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)


plt.subplot(5,2,8)
plt.ylabel('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.title("Agglomerative")
plotMultipleDims(ag_data_sil_k, colors, kVals, trial_num)
ax = plt.gca()
ax.set_ylim(0.0,1.0)
plt.legend(title = "k-value", loc=9, bbox_to_anchor=(1.2, 0.6))
```
